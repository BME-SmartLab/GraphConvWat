```python
import os
import pandas as pd
```

```python
df = pd.read_csv(os.path.join('..', 'experiments', 'Taylor_metrics.csv'), header=None)
df.head(3)
```

```python
df.columns = ['run_id', 'MSEc', 'sigma_pred']
df['sigma_true'] = 0
df['wds'] = [elem.split('-')[0] for elem in df['run_id']]
df['obs_rat'] = [elem.split('-')[2] for elem in df['run_id']]
df['model'] = [elem.split('-')[-1] for elem in df['run_id']]
df['num'] = [elem.split('-')[-2] for elem in df['run_id']]
df['adjacency'] = [elem.split('-')[3] for elem in df['run_id']]
```

```python
df.head(3)
```

```python
for wds in df['wds'].unique():
    sigma_true = df['sigma_pred'][(df['wds'] == wds) & (df['model'] == 'orig')].tolist()[0]
    df.loc[df['wds'] == wds, 'sigma_true'] = sigma_true
```

```python
df.head()
```

```python
df['corr_coeff'] = df['MSEc'] / (df['sigma_true']*df['sigma_pred'])
```

```python
df.head()
```

```python
df.to_csv(os.path.join('..', 'experiments', 'Taylor_metrics_processed.csv'))
```
