```python
import os
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(os.path.join('..', 'experiments', 'Taylor_metrics.csv'), header=None)
df.head(3)
```

```python
df.columns = ['run_id', 'MSEc', 'sigma_pred']
df['sigma_true'] = 0
df['seed'] = 0
df['wds'] = [elem.split('-')[0] for elem in df['run_id']]
df['placement'] = [elem.split('-')[1] for elem in df['run_id']]
df['obs_rat'] = [elem.split('-')[2] for elem in df['run_id']]
df['adjacency'] = [elem.split('-')[3] for elem in df['run_id']]
df['tag'] = [elem.split('-')[4] for elem in df['run_id']]
df['num'] = [elem.split('-')[5] for elem in df['run_id']]
df['model'] = [elem.split('-')[6] for elem in df['run_id']]
```

```python
df.head(3)
```

```python
for wds in df['wds'].unique():
    sigma_true = df['sigma_pred'][(df['wds'] == wds) & (df['model'] == 'orig')].tolist()[0]
    df.loc[df['wds'] == wds, 'sigma_true'] = sigma_true
```

```python
df.head()
```

```python
df['corr_coeff'] = df['MSEc'] / (df['sigma_true']*df['sigma_pred'])
```

### Handling results from sensor placement with random seeds

```python
seeds = np.array([1, 8, 5266, 739, 88867])
mask = df['placement'] == 'xrandom'
df.loc[mask, 'seed'] = seeds[df.loc[mask, 'num'].values.astype(int) % len(seeds)]
```

```python
df.to_csv(os.path.join('..', 'experiments', 'Taylor_metrics_processed.csv'))
```
